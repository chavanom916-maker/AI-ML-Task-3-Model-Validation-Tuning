```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load California Housing Dataset
housing = fetch_california_housing()

# Features and Target
X = housing.data
y = housing.target

# Create DataFrame
df = pd.DataFrame(X, columns=housing.feature_names)
df["Price"] = y

# Display first 5 rows
df.head()
```

```python
# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# Display the shapes of the datasets
print("Training Features Shape:", X_train.shape)
print("Testing Features Shape:", X_test.shape)
print("Training Target Shape:", y_train.shape)
print("Testing Target Shape:", y_test.shape)
```

```python
# Train Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
```

```python
# Predictions on training and testing data
train_predictions = dt_model.predict(X_train)
test_predictions = dt_model.predict(X_test)
```

```python
# Calculate RMSE
train_rmse = np.sqrt(mean_squared_error(y_train, train_predictions))
test_rmse = np.sqrt(mean_squared_error(y_test, test_predictions))

print("Training RMSE:", train_rmse)
print("Testing RMSE:", test_rmse)
```

```python
# Perform 5-Fold Cross Validation
cv_scores = cross_val_score(
    dt_model,
    X_scaled,
    y,
    cv=5,
    scoring="neg_root_mean_squared_error"
)

# Convert negative RMSE to positive RMSE
cv_rmse = -cv_scores

print("Cross Validation RMSE Scores:")
print(cv_rmse)

print("\nAverage Cross Validation RMSE:", cv_rmse.mean())
```

```python
# Define parameter grid
param_grid = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10]
}

# GridSearchCV
grid_search = GridSearchCV(
    estimator=DecisionTreeRegressor(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='neg_root_mean_squared_error'
)

# Train GridSearchCV
grid_search.fit(X_train, y_train)
```

```python
# Best Parameters
print("Best Parameters:", grid_search.best_params_)
```

```python
# Get the best model
best_model = grid_search.best_estimator_
```

```python
# Predictions using the best model
best_predictions = best_model.predict(X_test)
```

```python
# Calculate RMSE and R² Score
best_rmse = np.sqrt(mean_squared_error(y_test, best_predictions))
best_r2 = r2_score(y_test, best_predictions)

print("Optimized Model RMSE:", best_rmse)
print("Optimized Model R² Score:", best_r2)
```

```python
# Train Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Train Ridge Regression
ridge_model = Ridge()
ridge_model.fit(X_train, y_train)
```

```python
# Linear Regression
linear_predictions = linear_model.predict(X_test)
linear_rmse = np.sqrt(mean_squared_error(y_test, linear_predictions))
linear_r2 = r2_score(y_test, linear_predictions)

# Ridge Regression
ridge_predictions = ridge_model.predict(X_test)
ridge_rmse = np.sqrt(mean_squared_error(y_test, ridge_predictions))
ridge_r2 = r2_score(y_test, ridge_predictions)

# Display Results
print("Linear Regression RMSE:", linear_rmse)
print("Linear Regression R²:", linear_r2)

print("\nRidge Regression RMSE:", ridge_rmse)
print("Ridge Regression R²:", ridge_r2)

print("\nDecision Tree (Tuned) RMSE:", best_rmse)
print("Decision Tree (Tuned) R²:", best_r2)
```

```python
# Model Comparison Table
comparison = pd.DataFrame({
    "Model": ["Linear Regression", "Ridge Regression", "Decision Tree (Tuned)"],
    "RMSE": [linear_rmse, ridge_rmse, best_rmse],
    "R² Score": [linear_r2, ridge_r2, best_r2]
})

comparison
```

Final Model Selection Justification
1. Why this model was selected

After comparing all three models, the Tuned Decision Tree Regressor performed the best. It achieved the lowest RMSE (0.645430) and the highest R² Score (0.682099), which means it predicted house prices more accurately than Linear Regression and Ridge Regression.

2. How overfitting was reduced

At first, the Decision Tree model was clearly overfitting because it performed extremely well on the training data but not as well on the testing data. To improve this, GridSearchCV was used to tune the max_depth and min_samples_split parameters. This helped reduce overfitting and made the model perform better on unseen data.

3. Why the cross-validation results are trusted

Instead of depending on just one train-test split, 5-fold Cross Validation was used to evaluate the model. The average Cross Validation RMSE was 0.896157, which gives a more reliable idea of how the model is likely to perform on different datasets.

4. Trade-offs between simplicity and performance

Linear Regression and Ridge Regression are simpler models and easier to understand, but their prediction accuracy was lower. The Tuned Decision Tree is slightly more complex, but it produced better results. Based on the overall performance, it was the most suitable model for this task.
